Recreating figures 17.9, 17.10

```python
Psi = np.array([[0.95, 0.5],
                [0.1, 0.9]])
B = np.array([[1/6, 1/10],
              [1/6, 1/10],
              [1/6, 1/10],
              [1/6, 1/10],
              [1/6, 1/10],
              [1/6, 5/10]])
pi = np.array([2/3, 1/3])
dishonest_casino = HiddenMarkovModel(2, 6, Psi=Psi, B=B, pi=pi)
```

Show that, with enough samples, we can recover the parameters (up to permutation in rows/columns).

```python
samples = DishonestCasino.sample([100]*100)
```
