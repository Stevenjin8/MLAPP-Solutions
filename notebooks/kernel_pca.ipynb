```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets

from mlapp_models.kernel_pca import KernelPCA
from mlapp_models.utils import pairwise_rbf_kernel
```

# Kernel Principal Component Analysis

Section 14.4.4

## Create Data

```python
n_samples = 800
X, y = datasets.make_circles(n_samples=n_samples, factor=0, noise=0.1)
```

```python
plt.plot(*X[y == 0].T, ".", alpha=0.5)
plt.plot(*X[y == 1].T, ".", alpha=0.5)
```

## Apply Kernel PCA

```python
model = KernelPCA(data=X, kernel=pairwise_rbf_kernel)
```

```python
plt.plot(*model.predict(X[y == 0]).T, ".", alpha=0.4)
plt.plot(*model.predict(X[y == 1]).T, ".", alpha=0.4)
```

We can now linearly separate the data using just the first principle component.
